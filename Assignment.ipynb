```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the global population dataset
url = 'https://datahub.io/core/population/r/population.csv'
df = pd.read_csv(url)

# Show first few rows


print(df.head())

```

```python

# Filter data for specific countries
countries = ['Ghana', 'Nigeria', 'India', 'China', 'United States']
df_sel = df[df['Country Name'].isin(countries)]

# Line plot: Population growth over time
plt.figure(figsize=(10,6))
sns.lineplot(data=df_sel, x='Year', y='Value', hue='Country Name')
plt.title('Population Growth (1960–2020)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()

```

```python
# Bar chart: Population in 2020 for selected countries
df_2020 = df[df['Year'] == 2020]
df_2020_sel = df_2020[df_2020['Country Name'].isin(countries)]

plt.figure(figsize=(8,5))
sns.barplot(
    data=df_2020_sel,
    x='Country Name',
    y='Value',
    hue='Country Name',   # 👈 Added to remove FutureWarning
    palette='viridis',
    legend=False          # 👈 Hides redundant legend
)
plt.title('Population by Country (2020)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

```python

# Heatmap: Correlation example (though only 'Value' is numeric)
corr = df_sel[['Year', 'Value']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Year vs Population)')
plt.show()
```
